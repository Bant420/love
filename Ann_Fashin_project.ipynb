- Then we develop the model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, Activation
from tensorflow.keras.initializers import HeNormal
```

```python
# we initialize the model
model = Sequential()


#### Input Layer####
model.add(Flatten(input_shape=(28,28)))  # there is input shape is 28, 28 becaues is alredy given in the image data shape

##### Hidden Layer 1 ####
model.add(Dropout((0.2)))   # then useing these we remove the 20% neurons from hidden layer
model.add(Dense(128, kernel_initializer=HeNormal()))  # logits
        # Then we apply Batchnormalizationn
model.add(BatchNormalization())
        # activation --> Which shows the nonlinearity in function
model.add(Activation('relu'))



##### Hidden Layer 2 #####
model.add(Dropout((0.2)))
model.add(Dense(128, kernel_initializer=HeNormal()))
      # again we apply batch normalization
model.add(BatchNormalization())
      # and add the cativation function
model.add(Activation('relu'))


#### Output layer #####
model.add(Dense(10, activation = 'softmax', kernel_initializer = HeNormal()))
```

```python
# Then we check the model summary
model.summary()
```

- THen we compile the model

  - as using the optimiser as adam

  - metrics : accuracy

  - then the output is not binary then we shoose the loss is croosentropy

```python
from re import VERBOSE
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

- Then fit the data into the model

  - training data

  - epoch

  - batch size

  - and validation

```python
help(model.fit)
```

```python
history = model.fit(x_train,
                    y_train,
                    epochs = 50,
                    batch_size = 64,
                    validation_split = 0.2)
```

```python

#  here we have data of 60000 but we take 20% into validation
#  but then 48000 datat is remain
# in that we take batch size 64 then 48000 / 64 = 750 batch
#  if we dont use the validation batch then batches are 60000 / 64 = 938 batches
```

- Then we creaet the dataframe which have max_porb, index, prediction class, gt, and output

```python
max_prob = [np.max(i) for i in y_pred]
index = [np.argmax(i) for i in y_pred]
prediction_class = [class_names[i] for i in index]
Ground_Truth_class = [class_names[i] for i in y_test]



d1 = pd.DataFrame(zip(max_prob, index, prediction_class, Ground_Truth_class),
                  columns = ['Max_pob', 'Index', 'Prediction_class', 'Ground_Truth_class'])

#  then we pass the condition into it

con = d1['Prediction_class'] == d1['Ground_Truth_class'] # these conditon we pass into it

d1['Output'] = np.where(con, 1, 0)  # here if condition is corect then in output is 1 if not then 0

accuracy = d1['Output'].sum()/len(d1['Output'])
accuracy
```
